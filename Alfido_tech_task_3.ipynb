```python
import pandas as pd

# Load the dataset to inspect it
file_path = '/content/comments.csv'
df = pd.read_csv(file_path)

# Check the first few rows of the dataset
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
import nltk

# Initialize the VADER sentiment analyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

# Cleaning the comments by keeping the necessary column
comments = df['comment'].astype(str)

# Function to get sentiment using VADER
def get_sentiment_vader(text):
    score = sid.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'Positive'
    elif score['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment analysis to the comments
df['Sentiment'] = comments.apply(get_sentiment_vader)

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.countplot(df['Sentiment'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Show the first few rows with sentiment
df[['comment', 'Sentiment']].head()
```
